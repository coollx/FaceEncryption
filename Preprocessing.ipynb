```python
import pandas as pd
import os

root = 'CelebA_HQ_facial_identity_dataset'
# get all the folder name under the directory
def get_folder_name(path):
    folder_name = []
    for root, dirs, files in os.walk(path):
        for dir in dirs:
            folder_name.append(dir)
    return folder_name

id_list = get_folder_name(root + '/train')


id2name = pd.DataFrame(columns=['id', 'name'])

```

```python
# hq_A_mapping = pd.read_csv(root + '/CelebA-HQ-to-CelebA-mapping.txt', sep=' ', header=None, names = ['idx', 'orig_idx', 'orig_file'],on_bad_lines='skip')

#create a dictionary to map different type of CelebA
hq_A_mapping = {}
with open(root + '/CelebA-HQ-to-CelebA-mapping.txt', 'r') as f:
    for line in f:
        line = line.split()
        hq_A_mapping[line[0]] = line[2]
        
hq_A_mapping.pop('idx')
```

```python
id_name_mapping = {}

#create a dictionary to map name and id
with open(root + '/list_identity_celeba.txt', 'r') as f:
    for line in f:
        line = line.split()
        id_name_mapping[line[0]] = line[1]

id_name_mapping.pop('image_id')
```

```python
id_name_mapping[hq_A_mapping['5']]
```

```python
for id in id_list:
    #get the first image in the folder 
    img_name = os.listdir(root + '/train/' + id)[0]
    name = id_name_mapping[hq_A_mapping[img_name[:-4]]]
    
    #rename the folder with the name
    os.rename(root + '/train/' + id, root + '/train/' + name)
```

```python
for id in id_list:
    #get the first image in the folder 
    img_name = os.listdir(root + '/test/' + id)[0]
    name = id_name_mapping[hq_A_mapping[img_name[:-4]]]
    
    #rename the folder with the name
    os.rename(root + '/test/' + id, root + '/test/' + name)
```
